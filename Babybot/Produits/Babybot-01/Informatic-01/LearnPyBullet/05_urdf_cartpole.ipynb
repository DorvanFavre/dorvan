```python

import os
import time
import pybullet as p
import pybullet_data
import math
import numpy as np

timeStep = 1/30

p.connect(p.GUI)
p.setAdditionalSearchPath(pybullet_data.getDataPath())
p.setGravity(0,0,-9.81)
p.setRealTimeSimulation(0)
p.setTimeStep(timeStep)

camera_target_position = [0, 0, 0] # Position to look at
camera_distance = 2 # Distance from the target
camera_yaw = 60 # Yaw angle
camera_pitch = -40 # Pitch angle

# Calculate the camera position
camera_position = [
camera_target_position[0] + camera_distance * math.sin(math.radians(camera_yaw)) * math.cos(math.radians(camera_pitch)),
camera_target_position[1] + camera_distance * math.sin(math.radians(camera_pitch)),
camera_target_position[2] + camera_distance * math.cos(math.radians(camera_yaw)) * math.cos(math.radians(camera_pitch))
]
# Set the camera
p.resetDebugVisualizerCamera(cameraDistance=camera_distance,
cameraYaw=camera_yaw,
cameraPitch=camera_pitch,
cameraTargetPosition=camera_target_position)

p.resetSimulation()

plane_id = p.loadURDF("plane.urdf")
cartpole = p.loadURDF("cartpole.urdf",useFixedBase = 1,basePosition=[0.0, 0.0, 0.0], baseOrientation=[0, 0, 0.7071, 0.7071])
time.sleep(0.5)

p.changeDynamics(cartpole, -1, linearDamping=0, angularDamping=0)
p.changeDynamics(cartpole, 0, linearDamping=0, angularDamping=0)
p.changeDynamics(cartpole, 1, linearDamping=0, angularDamping=0)
# p.setJointMotorControl2(cartpole, 1, p.VELOCITY_CONTROL, force=0)
# p.setJointMotorControl2(cartpole, 0, p.TORQUE_CONTROL, force=0)
randstate = np.random.uniform(low=-0.5, high=0.5, size=(2,))
p.resetJointState(cartpole, 1, randstate[0])
p.resetJointState(cartpole, 0, randstate[1])

p.setJointMotorControl2(cartpole, 0, p.VELOCITY_CONTROL, 5, 0.1, force=1000)
#p.setJointMotorControl2(cartpole, 1, p.VELOCITY_CONTROL, 4, 0.1)

time.sleep(0.5)
for _ in range(100):

    #p.setJointMotorControl2(cartpole, 0, p.TORQUE_CONTROL, force=10)
    p.stepSimulation()
    time.sleep(timeStep)

# print("\n----------------\n\n")

# position, orientation = p.getBasePositionAndOrientation(robot_id)
# print("Base position: ", position, "Base orientation: ", orientation)

# numberOfJoints = p.getNumJoints(robot_id)
# print("Number of joints: ", numberOfJoints)

# for i in range(numberOfJoints):
#     jointInfo = p.getJointInfo(robot_id, i)
#     print(f"Joint {i} info: ", jointInfo)

#p.setJointMotorControl2(robot_id, 0, p.POSITION_CONTROL, targetPosition=1.0,targetVelocity=1.0, positionGain=0.01, velocityGain=0.01)
#p.setJointMotorControl2(robot_id,0,p.VELOCITY_CONTROL, targetVelocity=1.0, velocityGain=0.01)

# # Control the joint using position control
# for _ in range(100):
#     p.setJointMotorControl2(robot_id, 
#                             jointIndex=1, 
#                             controlMode=p.VELOCITY_CONTROL, 
#                             targetVelocity=1,
#                             force=10.0)  # Apply a constant torque

#     # Step the simulation
#     p.stepSimulation()
#     time.sleep(1/30)


time.sleep(2)
p.disconnect()
```
