```python
# Constants for the leg dimensions
A_R = 0.015
A_L = 0.05
B_R = 0.015
B_L = 0.05
C_R = 0.015
C_L = 0.08
D_R = 0.015
D_L = 0.14
E_R = 0.02

EFFORT = 1000
INERTIA = 0
VELOCITY = 1000

# Function to generate leg XML
def generate_leg_xml(name, x, y, rpy):
    return f"""
  <link name="a{name}">
    <visual>
      <geometry>
        <cylinder radius="{A_R}" length="{A_L}"/>
      </geometry>
      <origin rpy="1.57 0 0" xyz="0 {A_L/2} 0" />
      <material name="red">
        <color rgba="1 0 0 1"/>
      </material>
    </visual>
    <collision>
      <geometry>
        <cylinder radius="{A_R}" length="{A_L}"/>
      </geometry>
      <origin rpy="1.57 0 0" xyz="0 {A_L/2} 0" />
    </collision>
    <inertial>
      <mass value="0.1"/>
      <inertia ixx="{INERTIA}" ixy="0.0" ixz="0.0" iyy="{INERTIA}" iyz="0.0" izz="{INERTIA}"/>
    </inertial>
  </link>

  <joint name="a{name}_to_body" type="fixed">
    <origin rpy="0 0 {rpy}" xyz="{x} {y} 0"/>
    <parent link="body"/>
    <child link="a{name}"/>
  </joint>

  <!-- B -->

  <link name="b{name}">
    <visual>
      <geometry>
        <cylinder radius="{B_R}" length="{B_L/2}"/>
      </geometry>
      <origin rpy="1.57 0 0" xyz="0 {B_L/2} 0" />
      <material name="green">
        <color rgba="0 1 0 1"/>
      </material>
    </visual>
    <collision>
      <geometry>
        <cylinder radius="{B_R}" length="{B_L/2}"/>
      </geometry>
      <origin rpy="1.57 0 0" xyz="0 {B_L/2} 0" />
    </collision>
    <inertial>
      <mass value="0.1"/>
      <inertia ixx="{INERTIA}" ixy="0.0" ixz="0.0" iyy="{INERTIA}" iyz="0.0" izz="{INERTIA}"/>
    </inertial>

  </link>

  <joint name="u{name}" type="revolute">
    <origin rpy="0 0 0" xyz="0 0.05 0"/>
    <parent link="a{name}"/>
    <child link="b{name}"/>
    <axis xyz="0 0 1"/>
    <limit effort="{EFFORT}" lower="-1.57" upper="1.57" velocity="{VELOCITY}"/>
  </joint>

  <link name="c{name}">
    <visual>
      <geometry>
        <cylinder radius="{C_R}" length="{C_L/2}"/>
      </geometry>
      <origin rpy="1.57 0 0" xyz="0 {C_L/2} 0" />
      <material name="cyan">
        <color rgba="0 1 1 1"/>
      </material>
    </visual>
    <collision>
      <geometry>
        <cylinder radius="{C_R}" length="{C_L/2}"/>
      </geometry>
      <origin rpy="1.57 0 0" xyz="0 {C_L/2} 0" />
    </collision>
    <inertial>
      <mass value="0.1"/>
      <inertia ixx="{INERTIA}" ixy="0.0" ixz="0.0" iyy="{INERTIA}" iyz="0.0" izz="{INERTIA}"/>
    </inertial>

  </link>

  <joint name="v{name}" type="revolute">
    <origin rpy="0 0 0" xyz="0 0.05 0"/>
    <parent link="b{name}"/>
    <child link="c{name}"/>
    <axis xyz="1 0 0"/>
    <limit effort="{EFFORT}" lower="-1.57" upper="1.57" velocity="{VELOCITY}"/>
  </joint>

  <link name="d{name}">
    <visual>
      <geometry>
        <cylinder radius="{D_R}" length="{D_L/2}"/>
      </geometry>
      <origin rpy="1.57 0 0" xyz="0 {D_L/2} 0" />
      <material name="jaune">
        <color rgba="1 1 0 1"/>
      </material>
    </visual>
    <collision>
      <geometry>
        <cylinder radius="{D_R}" length="{D_L/2}"/>
      </geometry>
      <origin rpy="1.57 0 0" xyz="0 {D_L/2} 0" />
    </collision>
    <inertial>
      <mass value="0.1"/>
      <inertia ixx="{INERTIA}" ixy="0.0" ixz="0.0" iyy="{INERTIA}" iyz="0.0" izz="{INERTIA}"/>
    </inertial>

  </link>

  <joint name="w{name}" type="revolute">
    <origin rpy="0 0 0" xyz="0 0.08 0"/>
    <parent link="c{name}"/>
    <child link="d{name}"/>
    <axis xyz="1 0 0"/>
    <limit effort="{EFFORT}" lower="-3.14159" upper="1.57" velocity="{VELOCITY}"/>
  </joint>

  <link name="e{name}">
    <visual>
      <geometry>
        <sphere radius="{E_R}"/>
      </geometry>
      <origin rpy="0 0 0" xyz="0 0 0" />
      <material name="noir">
        <color rgba="0 0 0 1"/>
      </material>
    </visual>
    <collision>
      <geometry>
        <sphere radius="{E_R}"/>
      </geometry>
      <origin rpy="0 0 0" xyz="0 0 0" />
    </collision>
    <inertial>
      <mass value="0.01"/>
      <inertia ixx="{INERTIA}" ixy="0.0" ixz="0.0" iyy="{INERTIA}" iyz="0.0" izz="{INERTIA}"/>
    </inertial>

  </link>

  <joint name="x{name}" type="fixed">
    <origin rpy="0 0 0" xyz="0 0.14 0"/>
    <parent link="d{name}"/>
    <child link="e{name}"/>
  </joint>
"""

# Generate the full robot XML
def generate_robot_xml():
    legs = [
        generate_leg_xml(1, 0, 0.08, 0),
        generate_leg_xml(2, 0.069, 0.04, -1.0472),
        generate_leg_xml(3, 0.069, -0.04, -2.0944),
        generate_leg_xml(4, 0.0, -0.08, -3.1416),
        generate_leg_xml(5, -0.0693, -0.04, 2.0944),
        generate_leg_xml(6, -0.0693, 0.04, 1.0472),
    ]
    
    robot_xml = f"""
<robot name="hexapod">

  <link name="body">
    <visual>
      <geometry>
        <cylinder radius="0.08" length="0.06"/>
      </geometry>
      <origin xyz="0 0 0"/>
      <material name="blue">
        <color rgba="0 0 1 1"/>
      </material>
    </visual>
    <collision>
      <geometry>
        <cylinder radius="0.08" length="0.06"/>
      </geometry>
      <origin xyz="0 0 0"/>
    </collision>
    <inertial>
      <mass value="0.01"/>
      <inertia ixx="{INERTIA}" ixy="0.0" ixz="0.0" iyy="{INERTIA}" iyz="0.0" izz="{INERTIA}"/>
    </inertial>
  </link>

""" + "\n".join(legs) + "\n</robot>"

    return robot_xml

# Write to an output file
output_file_path = "hexapod_2.urdf"
with open(output_file_path, "w") as file:
    file.write(generate_robot_xml())

print(f"URDF XML generated and saved to '{output_file_path}'")
```
